```python
df_notnan_district = df[~df["District"].isna()]
```

```python
df_nan_district = df_nan_district.sample(frac=1)
```

```python
df_notnan_district = df_notnan_district.sample(frac=1)
```

```python
nan_k_ids = list(df_nan_district["Kanoon_ID"].unique())
not_nan_k_ids = list(df_notnan_district["Kanoon_ID"].unique())
```

```python
nan_k_ids
```

```python
path_to_text = "/Users/shashanksingh/Desktop/IND_PROJ/water_pollution/Kanoon_html_new_with_air"
```

```python
list_no_loc_paths = []

for k in nan_k_ids:
    
    list_no_loc_paths.append(path_to_text+"/"+str(int(k))+".txt")
```

```python
list_loc_paths = []

for k in not_nan_k_ids:
    
    list_loc_paths.append(path_to_text+"/"+str(int(k))+".txt")
```

```python
import shutil
import os
import glob

no_loc = "/Users/shashanksingh/Desktop/air_pollution/samik_cases/no_loc"
loc = "/Users/shashanksingh/Desktop/air_pollution/samik_cases/loc"
```

```python
if not os.path.exists(no_loc):
    os.mkdir(no_loc)
if not os.path.exists(loc):
    os.mkdir(loc)
```

```python

for f in list_no_loc_paths:

    if (len(glob.glob(no_loc+"/*.txt"))<30):
        try:
            shutil.copy(f, no_loc)
        except:
            pass
    
for f in list_loc_paths:
    
    if(len(glob.glob(loc+"/*.txt"))<30):
        try:
            shutil.copy(f, loc)
        except:
            pass
    
```
