```python
#list_paths = []
import tqdm
list_text = []

for k in tqdm.tqdm(k_ids):
    
    path = path_to_text+"/"+str(int(k))+".html"
    
    f = open(path, "r").read()
    soup = BeautifulSoup(f, 'lxml')
    
    judgements_div = soup.findAll("div", {"class": "judgments"})
    
    list_text.append(judgements_div[0].text)
    
```

```python
df["text"] = list_text
```
